**Okay, Let's do the first task!!!!** \
Calculate the area under the curve where x is between 0 and 1 using the Riemann sum.

**What is Reimann sum?** \
    Reimann sum is a method that helps us to find the area under a curve approximately.\
    

Lets find the left hand side expression

```python
### upper and lower limit x -> 0 to 1
lower_limit = 0
upper_limit = 1

### No of rectangles between 0 and 1
n = 10000

###Delta value
delta = (upper_limit - lower_limit)/n
print(delta)
```

```python
#function to find the area of the curve using sum of rectangles within limit.
def find_area_limit(n, delta):
    area_rect = [math.sqrt(1-((x/n)**2)) * (delta) for x in range(1, n+1)]

    return sum(area_rect)
```

```python
area_lhs = find_area_limit(n, delta)
print(area_lhs)
```

**Let's calulate the same area using with the integral (RHS)**

Let's use some library like sympy to find the integral with limits


```python
###Functtion for our task
f = sp.sqrt(1-x**2)

#to tell that x is a symbol 
x = sp.symbols('x')

#Finding the area using integral method
sp.integrate(f, (x, lower_limit, upper_limit))
```

It's seems like LHS and RHS value are same. \
**It's seems like we managed to find the area of the curve :)**
